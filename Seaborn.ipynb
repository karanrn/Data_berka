```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
account = pd.read_csv('account.csv',sep=';')
card = pd.read_csv('card.csv', sep=';')
trans = pd.read_csv('trans.csv', sep=';')
client = pd.read_csv('client.csv', sep = ';')
order = pd.read_csv('order.csv', sep = ';')
loan = pd.read_csv('loan.csv', sep = ';')
disp = pd.read_csv('disp.csv', sep = ';')
district = pd.read_csv('district.csv', sep = ';')
```

```python
import seaborn as sn
# Scatter plot (linear regressor line by default)
sn.lmplot(x='A1', y='A11', data=district, fit_reg=False)
```

```python
# Violin plot of credit card
plt.figure(figsize=(25,20))
sn.set_style('whitegrid')
sn.violinplot(x='type', y='card_id', data=card, inner=None)
sn.swarmplot(x='type', y='card_id', data=card, color ='k',alpha=0.7)
```

```python
# Boxplot
plt.figure(figsize=(25,20))
sn.set_style('whitegrid')
sn.boxplot(x='type', y='card_id', data=card)
```

```python
# Heatmap on correlation between attributes
corr = loan.corr()

# Heatmap
sn.heatmap(corr)
```

```python
sn.distplot(district.A11)
```

```python
# Density plot
sn.kdeplot(district.A1,district.A11)
plt.xlim(0,None)
```
